```python
import pandas as pd
import re

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/data_for_eda/JD.csv')

# Define a regex pattern to match sentences that start with "Here's a list of tangible skills"
pattern = r"Here's a list of.*?:"

# Function to remove sentences matching the pattern from the skills column
def remove_matching_sentences(text):
    # Use regex to find and remove matching sentences
    cleaned_text = re.sub(pattern, '', str(text), flags=re.IGNORECASE).strip()
    return cleaned_text

# Apply the function to the skills column
df['Skills'] = df['Skills'].apply(remove_matching_sentences)

# Save the cleaned data back to a new CSV file
cleaned_file_path = '/content/drive/MyDrive/data_for_eda/JD_cleaned_skills.csv'
df.to_csv(cleaned_file_path, index=False)

print(f"Sentences removed and new file saved as '{cleaned_file_path}'.")
```

## Change the header name of hr label csv

```python
import pandas as pd

# Define the path to your CSV file
file_path = '/content/drive/MyDrive/data_for_eda/RESUME LABELS - WEB Developer.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Define the new header names for the first three columns
new_header_names = ['APPLICANT NAME', 'ACCEPT/REJECT', 'REASON']

# Get the current column names
current_columns = df.columns.tolist()

# Update only the first three column headers
updated_columns = new_header_names + current_columns[3:]

# Assign the updated column names to the DataFrame
df.columns = updated_columns

# Save the updated DataFrame back to the same CSV file
df.to_csv(file_path, index=False)

print(f'Header names updated in the file: {file_path}')
```

## Merge the hr_label with csv for EDA

csv_path: /content/drive/MyDrive/data_for_eda/cleaned_skills.csv

```python
import pandas as pd

# Define file paths
files_to_compare = [
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Android Developer (1).csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Corprate Finance Analyst.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Data Science.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Finance Analyst.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Finance Manager.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - HR Manager.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - NLP_LLMs.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Software Developer.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Systems Administrator.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Talent Acquisition .csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - Technical recruiter.csv',
    '/content/drive/MyDrive/data_for_eda/RESUME LABELS - WEB Developer.csv'
]
cleaned_skills_path = '/content/drive/MyDrive/data_for_eda/resumes_data_extraction_cleaned_skills.csv'
new_csv_path = '/content/drive/MyDrive/data_for_eda/resumes_data_extraction_updated_cleaned_skills.csv'

# Load the cleaned_skills CSV
cleaned_skills_df = pd.read_csv(cleaned_skills_path)

# Load all resume label CSV files and concatenate them
resume_dfs = [pd.read_csv(file) for file in files_to_compare]
resume_labels_df = pd.concat(resume_dfs, ignore_index=True)

# Perform the merge operation based on 'APPLICANT NAME' and 'Name'
merged_df = pd.merge(cleaned_skills_df, resume_labels_df,
                     left_on='Name', right_on='APPLICANT NAME',
                     how='left')

# Define the order of columns in the final output
final_columns = [
    'Name', 'Gender', 'Education', 'Last_Company', 'Email', 'LinkedIn',
    'Phone_Number', 'Nationality', 'Designation', 'Degrees', 'Years_of_Experience',
    'Skills', 'Path', 'APPLICANT NAME', 'ACCEPT/REJECT', 'REASON'
]

# Reorder columns if necessary
merged_df = merged_df[final_columns]

# Save the updated DataFrame to a new CSV file
merged_df.to_csv(new_csv_path, index=False)

print(f'Updated file saved to: {new_csv_path}')
```
