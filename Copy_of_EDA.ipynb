Formula:
Experience Match Rate = (Number of Candidates with Required Experience / Total Number of Candidates) * 100

Explanation:
Measures the percentage of candidates who meet or exceed the required experience.

```python
import pandas as pd
import re

# Load the CSV files
jd_file_path = '/content/drive/MyDrive/data_for_eda/JD_cleaned_skills.csv'
candidate_file_path = '/content/drive/MyDrive/data_for_eda/resumes_data_extraction_updated_cleaned_skills.csv'

jd_df = pd.read_csv(jd_file_path)
candidate_df = pd.read_csv(candidate_file_path)

# Extract relevant columns
jd_experience = jd_df[['Designation', 'Years_of_Experience']].set_index('Designation').to_dict()['Years_of_Experience']
candidate_experience = candidate_df[['Designation', 'Years_of_Experience']].set_index('Designation').to_dict()['Years_of_Experience']

print(jd_experience)
print(candidate_experience)
```

```python
import re

# Function to normalize years of experience
def normalize_experience(exp):
    if 'to' in exp:
        numbers = re.findall(r'\d+', exp)
        avg_years = (float(numbers[0]) + float(numbers[1])) / 2
        return avg_years
    elif '+' in exp or '-' in exp:
        numbers = re.findall(r'\d+', exp)
        return float(numbers[0])
    else:
        years = re.search(r'(\d+) years?', exp)
        months = re.search(r'(\d+) months?', exp)
        total_years = float(years.group(1)) if years else 0
        total_months = float(months.group(1)) / 12 if months else 0
        return total_years + total_months

# Function to calculate matching rate for each role
def calculate_matching_rate(job_exp, resume_exp, tolerance=1):
    job_exp_normalized = normalize_experience(job_exp)
    resume_exp_normalized = normalize_experience(resume_exp)

    if abs(job_exp_normalized - resume_exp_normalized) <= tolerance:
        return 1
    return 0


job_exp_data=jd_experience
resume_exp_data=aggregated_resume_degrees
# Calculate match rate across all roles
total_matches = 0
total_roles = len(job_exp_data)

for role, job_exp in job_exp_data.items():
    if role in resume_exp_data:
        resume_exp = resume_exp_data[role]
        total_matches += calculate_matching_rate(job_exp, resume_exp)
    else:
        print(f"No resume data found for role: {role}")

# Calculate and print the overall match rate
overall_match_rate = (total_matches / total_roles) * 100
print(f"Overall Match Rate: {overall_match_rate:.2f}%")

# Optionally, you can calculate match rates for each role individually
for role, job_exp in job_exp_data.items():
    if role in resume_exp_data:
        resume_exp = resume_exp_data[role]
        match = calculate_matching_rate(job_exp, resume_exp)
        print(f"Role: {role}, Match: {'Yes' if match else 'No'}")
```

Formula:
Most Required Skills = Count of Each Skill in Job Descriptions

Explanation:
Counts the occurrences of each skill in job descriptions to identify the most frequently required skills.

```python
import pandas as pd
from collections import Counter

# Load the CSV file
jd_file_path = '/content/drive/MyDrive/data_for_eda/JD_cleaned_skills.csv'
jd_df = pd.read_csv(jd_file_path)

# Extract the skills column and process it
skills_series = jd_df['Skills']

# Initialize a counter for skills
skill_counter = Counter()

# Process each skill list in the DataFrame
for skills in skills_series:
    # Split the skills by newline or comma and remove any extra spaces
    skills_list = [skill.strip() for skill in skills.split('\n') if skill.strip()]
    # Update the counter with the current list of skills
    skill_counter.update(skills_list)

# Convert the counter to a DataFrame for easier viewing
skills_count_df = pd.DataFrame(skill_counter.items(), columns=['Skill', 'Count'])

# Sort the DataFrame by count in descending order
skills_count_df = skills_count_df.sort_values(by='Count', ascending=False)
print(skills_count_df)
# Save the results to a CSV file
skills_count_file_path = '/content/drive/MyDrive/data_for_eda/most_required_skills.csv'
skills_count_df.to_csv(skills_count_file_path, index=False)

print(f"Most required skills saved to: {skills_count_file_path}")
```

## 14. 🚨Acceptance Rate by Degree👇

Formula:
Acceptance Rate by Degree = (Number of Accepted Candidates by Degree / Total Number of Candidates by Degree) * 100

Explanation:
Measures the acceptance rate for candidates with each degree.

```python
import pandas as pd

# Load the CSV files
candidates_file_path = '/content/drive/MyDrive/data_for_eda/resumes_data_extraction_updated_cleaned_skills_cleaned.csv'  # Example path
job_descriptions_file_path = '/content/drive/MyDrive/data_for_eda/JD_cleaned_skills.csv'

# Load candidate data
candidates_df = pd.read_csv(candidates_file_path)

# Load job description data
job_descriptions_df = pd.read_csv(job_descriptions_file_path)

# Example function to determine if a candidate is accepted based on their degree
def is_accepted(degree, required_degrees):
    return degree in required_degrees

# Count the total number of candidates by degree
total_candidates_by_degree = candidates_df['Degrees'].value_counts()

# Initialize a dictionary to count accepted candidates by degree
accepted_candidates_by_degree = {degree: 0 for degree in total_candidates_by_degree.index}

# Process job descriptions to get required degrees
required_degrees_list = job_descriptions_df['Degrees'].dropna().tolist()
required_degrees_set = set()

for degrees in required_degrees_list:
    required_degrees_set.update(degree.strip() for degree in degrees.split(','))

# Count the number of accepted candidates by degree
for index, row in candidates_df.iterrows():
    degree = row['Degrees']
    if is_accepted(degree, required_degrees_set):
        accepted_candidates_by_degree[degree] += 1

# Calculate the acceptance rate by degree
acceptance_rate_by_degree = {
    degree: (accepted_candidates_by_degree[degree] / total_candidates_by_degree[degree]) * 100
    for degree in total_candidates_by_degree.index
}

# Convert to DataFrame for easier viewing
acceptance_rate_df = pd.DataFrame(list(acceptance_rate_by_degree.items()), columns=['Degrees', 'Acceptance Rate (%)'])
print(acceptance_rate_df)
# Save the results to a CSV file
acceptance_rate_file_path = '/content/drive/MyDrive/data_for_eda/acceptance_rate_by_degree.csv'
acceptance_rate_df.to_csv(acceptance_rate_file_path, index=False)

print(f"Acceptance rates by degree saved to: {acceptance_rate_file_path}")
```

## 19 .Correlation Between Experience and Acceptance
Formula:
Correlation = Pearson Correlation Coefficient between Experience and Acceptance Status

Explanation:
Measures the relationship between years of experience and acceptance status.

```python
import pandas as pd
from scipy.stats import pearsonr

# Load the CSV file
file_path = '/content/drive/MyDrive/data_for_eda/resumes_data_extraction_updated_cleaned_skills_cleaned.csv'  # Update with your actual file path
df = pd.read_csv(file_path)

# Display first few rows to verify


# Function to convert 'Years_of_Experience' to numeric values
def convert_experience(experience_str):
    try:
        # Remove non-numeric characters and convert to float
        return float(experience_str.split()[0].replace('years', '').replace('year', ''))
    except ValueError:
        return None

# Convert 'Years_of_Experience' column to numeric
df['Years_of_Experience_numeric'] = df['Years_of_Experience'].apply(convert_experience)

# Encode 'ACCEPT/REJECT' as binary (Accept = 1, Reject = 0)
df['Acceptance_Binary'] = df['ACCEPT/REJECT'].apply(lambda x: 1 if x == 'Accept' else 0)

# Drop rows with missing values in 'Years_of_Experience_numeric' or 'Acceptance_Binary'
df_clean = df.dropna(subset=['Years_of_Experience_numeric', 'Acceptance_Binary'])

# Calculate Pearson Correlation Coefficient
correlation, _ = pearsonr(df_clean['Years_of_Experience_numeric'], df_clean['Acceptance_Binary'])

# Print the correlation coefficient
print(f"\nPearson Correlation Coefficient between Experience and Acceptance: {correlation:.2f}")
```

```python
import pandas as pd
import plotly.express as px

# Load the CSV file
file_path = '/content/drive/MyDrive/data_for_eda/resumes_data_extraction_updated_cleaned_skills_cleaned.csv'  # Update with your actual file path
df = pd.read_csv(file_path)

# Group by 'Last_Company' and count the number of accepted and total candidates
acceptance_counts = df.groupby('Last_Company')['ACCEPT/REJECT'].value_counts().unstack(fill_value=0)

# Calculate the acceptance rate by company
acceptance_counts['Acceptance Rate (%)'] = (
    (acceptance_counts.get('Accept', 0) /
     (acceptance_counts.get('Accept', 0) + acceptance_counts.get('Reject', 0))) * 100
)

# Reset index to convert the groupby object into a DataFrame
acceptance_rates = acceptance_counts.reset_index()

# Sort the DataFrame by 'Acceptance Rate (%)' in descending order
acceptance_rates = acceptance_rates.sort_values(by='Acceptance Rate (%)', ascending=False)

# Create a bar chart using Plotly
fig = px.bar(acceptance_rates,
             x='Last_Company',
             y='Acceptance Rate (%)',
             title='Acceptance Rate by Last Company',
             labels={'Last_Company': 'Company', 'Acceptance Rate (%)': 'Acceptance Rate (%)'},
             text='Acceptance Rate (%)')

# Customize the chart
fig.update_traces(texttemplate='%{text:.2f}%', textposition='outside')
fig.update_layout(xaxis_title='Company',
                  yaxis_title='Acceptance Rate (%)',
                  xaxis_tickangle=-45,  # Rotate x-axis labels for better readability
                  plot_bgcolor='rgba(0,0,0,0)')

# Show the chart
fig.show()
```
