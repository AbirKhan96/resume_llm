```python
import pandas as pd
import random
import re
from pathlib import Path
from docx import Document
import ast

# Function to sanitize text
def sanitize_text(text):
    # Remove all control characters except newline, carriage return, and tab
    sanitized = re.sub(r'[^\x09\x0A\x0D\x20-\x7E\xA0-\uD7FF\uE000-\uFFFD]', '', text)
    return sanitized

# Function to insert custom text
def insert_custom_text(doc, custom_text, pattern):
    # Assuming pattern is a placeholder like '{{custom}}' in the document where you want to insert custom text
    return doc.replace(pattern, custom_text)

# Load CSV files
df = pd.read_csv('/content/gdrive/MyDrive/faking_skills_0706.csv')
main_df = pd.read_csv('/content/gdrive/MyDrive/yearofexperience.csv')
udf=pd.read_csv('/content/gdrive/MyDrive/yearofexperience_.csv')

# Filter data for android developers
adrd_dev = df[df['domain'] == 'technical_recruiter']   #change here
skills = adrd_dev['skills'].tolist()
print(adrd_dev.head())
main_df.dropna(inplace=True)
# print(main_df.head())
# Process each document
import ast
for ind, row in main_df.iterrows():
    desig = row['Designation']

    if 'technical recruiter' not in desig.lower():     #change here
        continue
    avail_skills=row["Skills"].split()
    # avail_skills.append('abir')
    # print(avail_skills)
    # break
    print(f"Processing: {desig}")
    document_path = row['Path']
    document = Document(document_path)
    doc = ''

    # Read the document content
    for paragraph in document.paragraphs:
        doc += paragraph.text + '\n'

    # Add random skills and contexts
    for i in range(20):                                #change here
        skill = random.choice(skills)
        cntx = ast.literal_eval(df[df['skills'] == skill]['context'].values[0])[:-1]
        cntx = random.choice(cntx)
        doc = insert_custom_text(doc, cntx + '\n', pattern='{{custom}}')  # Adjust pattern as needed
        # udf.at[ind, "skills"]
        avail_skills.append(skill)
    # Sanitize and save the document
    sanitized_doc = sanitize_text(doc)
    new_document = Document()
    new_document.add_paragraph(sanitized_doc)
    udf.at[ind, "Skills"]=avail_skills

    output_path = f'/content/gdrive/MyDrive/Faking_resumes/faked_resume_bulk/Technical Recruiter/{Path(document_path).name}'
    new_document.save(output_path)
udf.to_csv('/content/gdrive/MyDrive/yearofexperience_.csv',index=False)
print("Processing complete.")
```
